```python
import json
with open('scrapeme.json','r') as f:
    with open('camin.json','a') as f2:
        j = json.load(f)
        j.remove({})
        for i in j:
            item = {'url':i['url'],'type':'pick'}
            line = json.dumps(dict(item)) + ",\n"
            f2.write(line)
```
